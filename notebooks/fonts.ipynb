```python
import matplotlib
import matplotlib.font_manager as fm

print(matplotlib.get_cachedir())

font_family = matplotlib.rcParams["font.family"]
print(font_family)

font = fm.findfont(fm.FontProperties(family=["sans-serif"]))
print(font)
```

```python
from IPython.core.display import HTML


def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(
        font=fontname
    )


code = "\n".join(
    [make_html(font) for font in sorted(set([f.name for f in fm.fontManager.ttflist]))]
)

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```
